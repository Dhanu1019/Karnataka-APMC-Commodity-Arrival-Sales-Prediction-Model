```python
model = joblib.load("outputs/best_model.pkl")
df    = pd.read_csv("data/cleaned_data.csv")
 
print("\n✅ Model loaded from  outputs/best_model.pkl")
```

```python
def predict_arrival(district, taluk, market, commodity, unit, month_num):
    """
    Predict commodity arrival (in Quintals or Numbers) for given inputs.
 
    Parameters
    ----------
    district   : str   e.g. "Hassan"
    taluk      : str   e.g. "Hassan"
    market     : str   e.g. "HASSAN"
    commodity  : str   e.g. "Maize"
    unit       : str   "Quintal", "Numbers", or "Thousands"
    month_num  : int   1 to 12 (Jan=1, Dec=12)
 
    Returns
    -------
    float : predicted arrival quantity
    """
    row = pd.DataFrame([{
        "District" : district,
        "Taluk"    : taluk,
        "Market"   : market,
        "Commodity": commodity,
        "Unit"     : unit,
        "Month_Num": month_num,
    }])
    log_pred = model.predict(row)[0]
    return round(np.expm1(log_pred), 2)
```

```python
print("\n📦 Example Predictions:\n")
 
examples = [
    ("Hassan",    "Hassan",    "HASSAN",     "Maize",      "Quintal", 3),
    ("Mysore",    "Mysore",    "MYSORE",     "Paddy",      "Quintal", 8),
    ("Belgaum",   "Belgaum",   "BELGAUM",    "Groundnut",  "Quintal", 11),
    ("Bagalakot", "Badami",    "BADAMI",     "Sunflower",  "Quintal", 5),
    ("Kolar",     "Kolar",     "KOLAR",      "Tomato",     "Quintal", 7),
    ("Shimoga",   "Shimoga",   "SHIMOGA",    "Arecanut",   "Quintal", 2),
    ("Davanagere","Davanagere","DAVANAGERE", "Cotton",     "Quintal", 10),
    ("Udupi",     "Udupi",     "UDUPI",      "Coconut",    "Thousands",6),
]
 
print(f"  {'District':<14} {'Commodity':<12} {'Month':>5}  {'Predicted Arrival':>18}")
print("  " + "-" * 55)
month_names = {1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",
               7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"}
 
for district, taluk, market, commodity, unit, month_num in examples:
    pred = predict_arrival(district, taluk, market, commodity, unit, month_num)
    print(f"  {district:<14} {commodity:<12} {month_names[month_num]:>5}  "
          f"{pred:>15,.2f} {unit}")
```
