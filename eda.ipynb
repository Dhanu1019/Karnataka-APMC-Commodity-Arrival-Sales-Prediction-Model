```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
 
os.makedirs("outputs", exist_ok=True)
sns.set_style("whitegrid")
BLUE   = "#2563eb"
ORANGE = "#f97316"
GREEN  = "#16a34a"
RED    = "#dc2626"
 
df = pd.read_csv("data/cleaned_data.csv")
print("=" * 55)
print("STEP 2 : Exploratory Data Analysis")
print("=" * 55)
print(f"\n✅ Loaded cleaned data  →  {df.shape[0]} rows")
```

```python
top_comm = (
    df.groupby("Commodity")["Arrival"]
    .sum()
    .sort_values(ascending=True)
    .tail(15)
)
 
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(top_comm.index, top_comm.values, color=BLUE)
ax.set_xlabel("Total Arrival (all months & markets)", fontsize=11)
ax.set_title("Top 15 Commodities by Total Arrival — Karnataka APMC 2012",
             fontsize=13, fontweight="bold")
ax.bar_label(bars, fmt="{:,.0f}", padding=4, fontsize=8)
plt.tight_layout()
plt.savefig("outputs/eda_top_commodities.png", dpi=150)
plt.close()
print("✅ Saved  →  outputs/eda_top_commodities.png")
```

```python
month_order = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]
monthly = df.groupby("Month")["Arrival"].sum().reindex(month_order)
 
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(monthly.index, monthly.values, marker="o",
        color=ORANGE, linewidth=2.5, markersize=7)
ax.fill_between(monthly.index, monthly.values, alpha=0.15, color=ORANGE)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Total Arrival", fontsize=11)
ax.set_title("Monthly Commodity Arrival Trend — Karnataka APMC 2012",
             fontsize=13, fontweight="bold")
for i, (m, v) in enumerate(zip(monthly.index, monthly.values)):
    ax.annotate(f"{v/1e6:.1f}M", (m, v), textcoords="offset points",
                xytext=(0, 8), ha="center", fontsize=8)
plt.tight_layout()
plt.savefig("outputs/eda_monthly_trend.png", dpi=150)
plt.close()
print("✅ Saved  →  outputs/eda_monthly_trend.png")
```

```python
month_order = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]
monthly = df.groupby("Month")["Arrival"].sum().reindex(month_order)
 
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(monthly.index, monthly.values, marker="o",
        color=ORANGE, linewidth=2.5, markersize=7)
ax.fill_between(monthly.index, monthly.values, alpha=0.15, color=ORANGE)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Total Arrival", fontsize=11)
ax.set_title("Monthly Commodity Arrival Trend — Karnataka APMC 2012",
             fontsize=13, fontweight="bold")
for i, (m, v) in enumerate(zip(monthly.index, monthly.values)):
    ax.annotate(f"{v/1e6:.1f}M", (m, v), textcoords="offset points",
                xytext=(0, 8), ha="center", fontsize=8)
plt.tight_layout()
plt.savefig("outputs/eda_monthly_trend.png", dpi=150)
plt.close()
print("✅ Saved  →  outputs/eda_monthly_trend.png")
```

```python
top_dist = (
    df.groupby("District")["Arrival"]
    .sum()
    .sort_values(ascending=True)
    .tail(10)
)
 
fig, ax = plt.subplots(figsize=(10, 5))
ax.barh(top_dist.index, top_dist.values, color=GREEN)
ax.set_xlabel("Total Arrival", fontsize=11)
ax.set_title("Top 10 Districts by Total Commodity Arrival",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("outputs/eda_top_districts.png", dpi=150)
plt.close()
print("✅ Saved  →  outputs/eda_top_districts.png")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
 
axes[0].hist(df["Arrival"], bins=50, color=RED, edgecolor="white")
axes[0].set_xlabel("Arrival")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Arrival Distribution (raw)")
 
import numpy as np
axes[1].hist(np.log1p(df["Arrival"]), bins=50, color=BLUE, edgecolor="white")
axes[1].set_xlabel("log(Arrival + 1)")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Arrival Distribution (log scale — more normal)")
 
plt.suptitle("Why we use log transform for modeling", fontsize=12,
             fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("outputs/eda_distribution.png", dpi=150)
plt.close()
print("✅ Saved  →  outputs/eda_distribution.png")
```
